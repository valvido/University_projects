### Rješenje

Napomene.
- Gibanje lopte aproksimiramo gibanjem točke jednake mase, tj. zanemarujemo utjecaj rotacije lopte na gibanje.

- Gibanje je ravninsko

a) U ovom zadatku promatramo dvodimenzionalno gibanje. Radijvektor položaja lopte je
$$
\vec{r}(t) = (x(t), y(t), 0)=x(t)\vec{i}+y(t)\vec{j}
$$

Sile koje djeluju na tijelo su:

- Sila teže:  $\vec{G}=-mg\vec{j}$
- Sila otpora zraka:  $\vec{F_o}= -m\alpha v = -m\alpha\dot{x}\vec{i}-m\alpha\dot{y}\vec{j}$

Iz drugog Newtonovog aksioma slijedi:
$$
m\vec{a}=\vec{G}+\vec{F_o}
$$
$$
m\ddot{\vec{r}}(t)=-mg\vec{j}-m\alpha\dot{x}(t)\vec{i}-m\alpha\dot{y}(t)\vec{j}
$$

Raspisom po komponentama dobivamo:
$$
\vec{i}...m\ddot{x}=-m\alpha\dot{x}
$$
$$
\vec{j}...m\ddot{y}=-mg-m\alpha\dot{y}
$$

$$
\vec{i}...\ddot{x}=-\alpha\dot{x}
$$
$$
\vec{j}...\ddot{y}=-g-\alpha\dot{y}
$$

uz početni položaj i brzinu
$$
\vec{r}(0)=(x(0),y(0),0)=(0,0,0)
$$
$$
\vec{v}(0)=(\dot{x}(0),\dot{y}(0),0)=(v_0 cos\theta,v_0 sin\theta,0)
$$
gdje je $v_0=|\vec{v_0}|$

Rješavamo:
$$
\left\{
\begin{array}{l}
\ddot{x} = -\alpha \dot{x} \\
x(0) = 0 \\
\dot{x}(0) = v_0 \cos \theta 
\end{array}
\right.
$$



$$
\left\{
\begin{array}{l}
\ddot{y} = -\alpha \dot{y} - g \\
y(0) = 0 \\
\dot{y}(0) = v_0 \sin \theta 
\end{array}
\right.
$$

Rješavanjem sustava dobivamo
$$
x(t)=\frac{v_0cos\theta}{\alpha}(1-e^{-\alpha t})
$$
$$
\dot{x}(t)=v_0cos\theta \cdot e^{-\alpha t}
$$
$$
y(t)=(\frac{v_0sin\theta}{\alpha}+\frac{g}{\alpha^2})(1-e^{-\alpha t})-\frac{g}{\alpha}t
$$
$$
\dot{y}(t)=(v_0sin\theta+\frac{g}{\alpha})e^{-\alpha t}-\frac{g}{\alpha}
$$

### kut θ uz koji je x-koordinata točke A maksimalna

b) Pretpostavimo da je $\alpha = \frac{g}{v_0}$. Korištenjem a) djela imamo 
$$
y(t)=(\frac{v_0^2sin\theta}{g}+\frac{v_0^2}{g})(1-e^{-\frac{g}{v_0}t})-v_0t
$$
U najvišoj točci putanje vrijedi:
$$\dot{y}(t)=0$$
$$v_0e^{\frac{-g}{v_0}t}(sin\theta +1)-v_0=0$$



Iz gornje jednadžbe računamo vrijeme $t_A$, tj. vrijeme potrebno da lopta postigne maksimalnu visinu.
$$
t_A=\frac{v_0}{g}ln(sin\theta+1)
$$

uvrštavanjem u $x(t)$ dobivamo visinu najviše točke
$$
x(t_A)=\frac{v_0cos\theta}{\alpha}(1-e^{-\alpha \frac{v_0 ln(sin\theta+1)}{g}})
$$

Uvrštavanjem $\alpha=\frac{g}{v_0}$ i sređivanjem izraza imamo 
$$
x(t_A)=\frac{v_0^2 cos\theta}{g}(1-(sin\theta+1)^{-1})
$$
$$
x(t_A)=\frac{v_0^2cos\theta sin\theta}{g(sin(\theta)+1)}
$$

Problem svodimo na maksimizaciju 
$$
x(\theta)=\frac{v_0^2cos\theta sin\theta}{g(sin(\theta)+1)}
$$

Odnosno tražimo rješenje jednadžbe 
$$
\dot{x}(\theta)=0
$$


$$
\dot{x}(\theta) = \frac{v_0^2}{g} \frac{ (\cos^2 \theta - \sin^2 \theta)(\sin \theta + 1) - \cos \theta \sin \theta \cos \theta}{(\sin \theta + 1)^2}
$$
$$
\dot{x}(\theta) = \frac{ \left( \cos^2 \theta - \sin^2 \theta - \sin^2 \theta \sin \theta \right)}{(\sin \theta + 1)^2}=0
$$
$$
(\sin \theta + 1)^2\neq 0 => sin\theta \neq -1
$$
$$
 cos^2\theta - sin^2\theta - sin^3 \theta=0
$$

$$
 1 -2sin^2\theta - sin^3 \theta=0 => t=sin\theta
$$
$$
-t^3-2t^2+1=0 => t=\frac{-1\pm\sqrt{5}}{2}
$$
Kako je $\frac{-1-\sqrt{5}}{2}<-1$ slijedi da je jedino rjesenje 
$$
sin\theta=\frac{-1+\sqrt{5}}{2} => \theta = arcsin(\frac{-1+\sqrt{5}}{2})\approx 0.6623 
$$
$\theta \approx 0.6623$ rad, tj.
$\theta \approx 38.21$ stupnjeva

Grafički prikaz rješenja

```python
import numpy as np
import matplotlib.pyplot as plt

v0=10
g = 9.81  

def x(theta):
    return (v0**2*np.cos(theta)*(1-1/(np.sin(theta)+1)))/g

theta_values = np.linspace(0.01, np.pi/2, 500) #vrijednosti kuta theta mogu ici od 0 do pi/2

x_values = x(theta_values)
h_max_index = np.argmax(x_values)  


print(f"Maksimalni theta u radijanima: {theta_values[h_max_index]:.2f} ")
theta_deg = np.degrees(theta_values)
print(f"Maksimalni theta u stupnjevima: {theta_deg[h_max_index]:.2f} ")

plt.figure(figsize=(8, 6))
plt.plot(np.degrees(theta_values), x_values, label='x(theta)')
plt.xlabel('Theta (u stupnjevima)')
plt.ylabel('x(theta)')
plt.title('Graf od x(theta)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

v0=10
g = 9.81  

def x(theta):
    numerator = (v0**2 / g) * (np.cos(theta)**2 - np.sin(theta)**2 - np.sin(theta)**2 * np.sin(theta))
    denominator = (np.sin(theta) + 1)**2
    return numerator / denominator


theta_values = np.linspace(0.01, np.pi/2, 500)

x_values = x(theta_values)

plt.figure(figsize=(8, 6))
plt.plot(np.degrees(theta_values), x_values, label='x(theta)')
plt.xlabel('Theta (u stupnjevima)')
plt.ylabel('x(theta)')
plt.title('Graf od x\'(theta)')
plt.grid(True)
plt.legend()
plt.show()
```

Objašnjenje koda:
Numerički rješavamo zadani ODJ iz a) dijela te dobivamo, kao povratnu vrijednost funkcije odeint, matricu svih vrijednosti $[x,x',y,y']$  u određenom trenutku t. Zatim iz te matice promatramo sve vrijednosti od y i vracamo indeks najveće (to je y kordinata najvišu točku putanje lopte). Uz pomoć indeksa računamo y i x koordinatu te točke B.

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

g = 9.81 
v0 = 20  
theta = np.pi / 4
alpha = 0.5 

def rhs(p, t, alpha, g):
    x, dxdt, y, dydt = p  
    dot_p = [dxdt,
             -alpha * dxdt,
             dydt,
             -alpha * dydt - g]
    return dot_p

p0 = [0, v0 * np.cos(theta), 0, v0 * np.sin(theta)]

n=1000
a_0=0
a_1=5
t = np.linspace(a_0, a_1,n) 

sol = odeint(rhs, p0, t, args=(alpha,g)) #odeint je ekvivalent funckije lsode iz Octave


y_values = sol[:, 2]
t_max_index = np.argmax(y_values)  
x_max = sol[t_max_index, 0]  
y_max = y_values[t_max_index]  
h=a_1/n
t=a_0+t_max_index*h

print(f"Najviša točka B: ({x_max} m, {y_max} m)")
print(f"Index: {t_max_index:.2f}")
print(f"Vrijeme:{t}")

plt.figure(figsize=(10, 6))
plt.plot(sol[:, 0], sol[:, 2], label='Putanja lopte')
plt.scatter([x_max], [y_max], color='red', marker='o', label='Najviša točka B')
plt.xticks(np.arange(0, 30, 5))
plt.xlim(0, 30)
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Putanja lopte s najvišom točkom B')
plt.legend()
plt.grid(True)
plt.show()
```

### Određivanje kuta $\theta$

```python
Ako je $α = 0.2$ numerički odredite kut θ uz koji lopta pogada točku B (odredite sva rješenja).
```

Objasnjenje koda:
Definiram funkcije gibanja x(t,theta), y(t,theta) i funkciju koja mjeri udaljenost između lopte i točke B. Problem se svodi na minimizaciju te funkcije korištenjem minimize_scalar, tj. funkcije koja radi isto što i fminbnd


```python
import numpy as np
from scipy.optimize import minimize_scalar

g = 9.81
v0 = 20
alpha = 0.2
x_B = 11.85
y_B = 6.99

# Funkcije gibanja iz a) dijela zadatka u ovisnosti o t i theta
def x_t(t, theta):
    return (v0 * np.cos(theta) / alpha) * (1 - np.exp(-alpha * t))

def y_t(t, theta):
    return ((v0 * np.sin(theta) / alpha) + (g / alpha**2)) * (1 - np.exp(-alpha * t)) - (g / alpha) * t

def distance_to_target(theta):
    a_1 = 5
    a_0=0
    n=1000 #možemo smanjiti grešku povečavanjem broja intervala
    t = np.linspace(a_0, a_1, 1000)
    
    x = x_t(t, theta)
    y = y_t(t, theta)
    
    distances = np.sqrt((x - x_B)**2 + (y - y_B)**2)
    min_distance = np.min(distances)
    
    return min_distance

# minimize_scalar je ekvivalentno funkciji fminbnd iz Octave
result = minimize_scalar(distance_to_target, bounds=(0, np.pi/2), method='bounded')

optimal_theta = result.x
min_distance = result.fun

print(f'Optimalni kut θ: {optimal_theta} radijana ({np.degrees(optimal_theta)} stupnjeva)')
print(f'Greška: {min_distance} metara')
```
